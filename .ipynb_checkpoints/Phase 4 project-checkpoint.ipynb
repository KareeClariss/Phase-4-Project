#  Project Summary: Modeling Crash Severity and Contributory Causes in Chicago

This project applies supervised machine learning to the Chicago Traffic Crash dataset (~993,000 records) to convert raw traffic data into predictive, actionable insights for road safety stakeholders.



##  Business Problem and Context

Road traffic accidents in Chicago pose significant public health and economic challenges. The main problem is the analytical gap: stakeholders lack tools to determine which combinations of factors—such as driver behavior, environmental conditions, and roadway defects—most strongly influence severe crashes. Without this, designing preventative interventions and allocating resources optimally is difficult.

This project addresses the gap by building a robust classification model to predict the primary cause of accidents, enabling a transition from reactive to proactive safety planning.



## Core Objectives

1. **Model Development:** Create a high-performance classification model (XGBoost Classifier) to predict the Primary Contributory Cause of a crash.
2. **Factor Identification:** Assess the impact of conditions (speed limits, road defects, weather, etc.) on crash outcomes.
3. **Spatial and Temporal Analysis:** Map crash hotspots and identify critical peak hours/days to inform resource deployment.
4. **Actionable Strategy:** Provide evidence-based recommendations to city management for improving infrastructure, traffic control, and data collection.

---

##  Methodology and Data Preparation

###  Data Source and Cleaning

- **Source:** Chicago Traffic Crash dataset (~993k records).  
- **Preprocessing:** Removed columns with excessive missing data and records lacking essential features for modeling.

###  Feature Engineering & Balancing

- **Target Aggregation:** Reduced the original `prim_contributory_cause` (40+ categories) to 7 broader Cause Groups for better model interpretability and convergence:
  - Speed & Right-of-Way Failure  
  - Impairment & Recklessness  
  - Maneuver & Positioning Error  
  - Disregard Traffic Controls  
  - Distraction & Visibility  
  - External/Environmental/Mechanical  
  - Unclassified/Administrative
- **Class Imbalance Mitigation:** Applied Random OverSampling (ROS) to balance target classes, allowing the model to learn from rare but critical severe crashes.
- **Data Transformation:** Encoded categorical features and scaled numerical features using `StandardScaler`.



##  Modeling Performance and Findings

###  Model Evaluation

|------------------------|-----------------|----------------------------------------------------------------------|

###  Key Predictive Insights

- **Top Features:** Roadway defects, weather conditions, and time of day are the most influential predictors of crash causes.
- **Spatial/Temporal Trends:** Identified high-frequency hotspots and peak crash periods, guiding targeted interventions.

# DATA UNDERSTANDING
This data was derived from the [Chicago Data Portal](https://data.cityofchicago.org/Transportation/Traffic-Crashes-Vehicles/68nd-jvt3/about_data). This dataset contains information about vehicles (or units as they are identified in crash reports) involved in a traffic crash. The data has approximately 993k rows and 48 columns. After cleaning and dropping irrelevant columns, this analysis will be using the following columns:

* `crash_date` – The date and time when the crash occurred.

* `posted_speed_limit` – The speed limit (in mph) posted at the crash location.

* `traffic_control_device` – The type of traffic control device present at the crash site (e.g., stop sign, traffic signal).

* `device_condition` – The condition of the traffic control device at the time of the crash.

* `weather_condition` – The weather condition during the crash (e.g., clear, rain, snow).

* `lighting_condition` – The lighting condition at the time of the crash (e.g., daylight, dark – no streetlights).

* `first_crash_type` – The type of initial impact or collision in the crash.

* `trafficway_type` – The design or type of roadway where the crash occurred (e.g., one-way, divided highway).

* `alignment` – The road alignment where the crash occurred (e.g., straight, curve).

* `road_defect` – Any reported defect in the road that may have contributed to the crash.

* `crash_type` – A general classification of the crash (e.g., rear-end, sideswipe).

* `date_police_notified` – The date when the police were notified about the crash.

* `prim_contributory_cause` – The primary cause determined to have contributed to the crash.

* `sec_contributory_cause` – A secondary factor contributing to the crash.

* `street_no` – The street number where the crash occurred.

* `street_direction` – The compass direction (e.g., N, S, E, W) of the street where the crash occurred.

* `street_name` – The name of the street where the crash happened.

* `num_units` – The number of vehicles or units involved in the crash.

* `most_severe_injury` – The most serious injury outcome from the crash (e.g., fatal, no injury).

* `injuries_total` – The total number of injuries reported in the crash.

* `latitude` – The geographic latitude coordinate of the crash location.

* `longitude` – The geographic longitude coordinate of the crash location.

* `location` – A combined geographic point (latitude and longitude) representing the crash site.

## Grouping Primary Contributory Causes

To simplify analysis, the `prim_contributory_cause` column was mapped into broader categories using the `cause_groups` dictionary. This reduces over 40 unique causes into 7 meaningful groups:

1. **Speed & Right-of-Way Failure:** Violations related to yielding, tailgating, speeding, and improper turns.  
2. **Maneuver & Positioning Error:** Includes improper passing, lane usage, turning, backing, and driving skill issues.  
3. **Disregard Traffic Controls:** Ignoring traffic signals, stop signs, yield signs, road markings, and school bus rules.  
4. **Impairment & Recklessness:** Incidents due to alcohol, drugs, reckless driving, or physical impairment.  
5. **Distraction & Visibility:** Distractions inside/outside the vehicle, cellphone use, texting, or obscured vision.  
6. **External/Environmental/Mechanical:** Weather, road defects, vehicle condition, animals, and other external hazards.  
7. **Unclassified/Administrative:** Cases labeled *NOT APPLICABLE* or *UNABLE TO DETERMINE*.

The mapping is applied to create a new column `Cause_Group` in the dataframe. Using `value_counts()`, the distribution of these categories can be analyzed to identify major risk factors and trends in traffic incidents.


```python
cause_groups = {
    # 1. Right-of-Way & Speed Violations (High Volume, Focus on Flow/Priority)
    'FAILING TO YIELD RIGHT-OF-WAY': 'Speed & Right-of-Way Failure',
    'FOLLOWING TOO CLOSELY': 'Speed & Right-of-Way Failure',
    'FAILING TO REDUCE SPEED TO AVOID CRASH': 'Speed & Right-of-Way Failure',
    'EXCEEDING SAFE SPEED FOR CONDITIONS': 'Speed & Right-of-Way Failure',
    'EXCEEDING AUTHORIZED SPEED LIMIT': 'Speed & Right-of-Way Failure',
    'TURNING RIGHT ON RED': 'Speed & Right-of-Way Failure',

    # 2. Basic Maneuver Errors (Skill, Positioning, and Attention)
    'IMPROPER OVERTAKING/PASSING': 'Maneuver & Positioning Error',
    'IMPROPER BACKING': 'Maneuver & Positioning Error',
    'IMPROPER LANE USAGE': 'Maneuver & Positioning Error',
    'DRIVING SKILLS/KNOWLEDGE/EXPERIENCE': 'Maneuver & Positioning Error',
    'IMPROPER TURNING/NO SIGNAL': 'Maneuver & Positioning Error',
    'DRIVING ON WRONG SIDE/WRONG WAY': 'Maneuver & Positioning Error',

    # 3. Disregard of Traffic Controls/Signs (Deliberate Violation)
    'DISREGARDING TRAFFIC SIGNALS': 'Disregard Traffic Controls',
    'DISREGARDING STOP SIGN': 'Disregard Traffic Controls',
    'DISREGARDING OTHER TRAFFIC SIGNS': 'Disregard Traffic Controls',
    'DISREGARDING ROAD MARKINGS': 'Disregard Traffic Controls',
    'DISREGARDING YIELD SIGN': 'Disregard Traffic Controls',
    'PASSING STOPPED SCHOOL BUS': 'Disregard Traffic Controls',

    # 4. Impairment & Reckless Driving
    'OPERATING VEHICLE IN ERRATIC, RECKLESS, CARELESS, NEGLIGENT OR AGGRESSIVE MANNER': 'Impairment & Recklessness',
    'UNDER THE INFLUENCE OF ALCOHOL/DRUGS (USE WHEN ARREST IS EFFECTED)': 'Impairment & Recklessness',
    'HAD BEEN DRINKING (USE WHEN ARREST IS NOT MADE)': 'Impairment & Recklessness',
    'PHYSICAL CONDITION OF DRIVER': 'Impairment & Recklessness',

    # 5. Distraction & Visibility Issues
    'DISTRACTION - FROM INSIDE VEHICLE': 'Distraction & Visibility',
    'DISTRACTION - FROM OUTSIDE VEHICLE': 'Distraction & Visibility',
    'CELL PHONE USE OTHER THAN TEXTING': 'Distraction & Visibility',
    'TEXTING': 'Distraction & Visibility',
    'DISTRACTION - OTHER ELECTRONIC DEVICE (NAVIGATION DEVICE, DVD PLAYER, ETC.)': 'Distraction & Visibility',
    'VISION OBSCURED (SIGNS, TREE LIMBS, BUILDINGS, ETC.)': 'Distraction & Visibility',

    # 6. External, Environmental & Mechanical Factors
    'WEATHER': 'External/Environmental/Mechanical',
    'EQUIPMENT - VEHICLE CONDITION': 'External/Environmental/Mechanical',
    'ROAD ENGINEERING/SURFACE/MARKING DEFECTS': 'External/Environmental/Mechanical',
    'ROAD CONSTRUCTION/MAINTENANCE': 'External/Environmental/Mechanical',
    'EVASIVE ACTION DUE TO ANIMAL, OBJECT, NONMOTORIST': 'External/Environmental/Mechanical',
    'ANIMAL': 'External/Environmental/Mechanical',
    'RELATED TO BUS STOP': 'External/Environmental/Mechanical',
    'OBSTRUCTED CROSSWALKS': 'External/Environmental/Mechanical',
    'BICYCLE ADVANCING LEGALLY ON RED LIGHT': 'External/Environmental/Mechanical',
    'MOTORCYCLE ADVANCING LEGALLY ON RED LIGHT': 'External/Environmental/Mechanical',
    
    # 7. Unclassified/Administrative (Left separate as requested)
    'NOT APPLICABLE': 'Unclassified/Administrative',
    'UNABLE TO DETERMINE': 'UNABLE TO DETERMINE' 
}

df['Cause_Group'] = df['prim_contributory_cause'].map(cause_groups)
print(df['Cause_Group'].value_counts())
print(df.columns)
```

## Interactive Table of Primary Contributory Cause Groups

This code creates an interactive table using Plotly to display the mapping of individual traffic incident causes into broader groups:

   - `header`: Defines column names, background color, font size, and alignment.  
   - `cells`: Populates the table with grouped causes, alternating background colors for readability, and aligns text.  
4. **Layout Customization:** Adds a centered title, adjusts margins, and sets table height for better display.  

The resulting interactive table allows users to quickly explore which primary causes belong to each broader category, providing a clear and concise overview of traffic incident patterns.

```python

# Then import the required libraries
import plotly.graph_objects as go
import pandas as pd  # Also adding pandas import which was missing

# Convert dictionary to DataFrame
cause_df = pd.DataFrame(list(cause_groups.items()), columns=['Cause', 'Group'])
grouped_table = cause_df.groupby('Group')['Cause'].apply(lambda x: ', '.join(x)).reset_index()

# Create Plotly table
fig = go.Figure(data=[go.Table(
    header=dict(
        values=["<b>Group</b>", "<b>Causes</b>"],
        fill_color="#1f77b4",
        font=dict(color='white', size=14),
        align='left'
    ),
    cells=dict(
        values=[grouped_table['Group'], grouped_table['Cause']],
        fill_color=[['#f9f9f9', '#ffffff'] * len(grouped_table)],
        align='left',
        font=dict(size=12)
    )
)])

fig.update_layout(
    title="Primary Contributory Cause Groups – Key Table",
    title_x=0.4,
    margin=dict(l=10, r=10, t=60, b=10),
    height=600
)

fig.show()
```

### `Injury Severity` Distribution

###  `Top 5 Contributory Causes of Crashes` Distribution

We will look at the top 5 most common causes of crashes and how many crashes have occured due to each particular reason

```python
# Get top 5 contributory causes
top_5 = df['prim_contributory_cause'].value_counts().nlargest(5).index

# Filter dataset to include only top 5
df_top5 = df[df['prim_contributory_cause'].isin(top_5)]

# Create vertical countplot
plt.figure(figsize=(10,6))
sns.countplot(
    x='prim_contributory_cause',
    data=df_top5,
    order=top_5,
    palette='Set2'   
)

plt.title('Top 5 Primary Contributory Causes of Crashes', fontsize=14, weight='bold')
plt.xlabel('Primary Contributory Cause', fontsize=12)
plt.ylabel('Number of Crashes', fontsize=12)
plt.xticks(rotation=30, ha='right')  
plt.tight_layout()
plt.show()
```

As we can see in the countplot, more than 350,000 crashes were left undetermined this may suggest data limitations. Another major contributor is "Failing to yield Right-of-Way" which is often linked to driver error or violation of rules. "Following  Too Closely" and "Improper Passing" have similar number of crashes and also occur during congested traffic or distracted driving. Lastly some causes are "Not Applicable" where the cases were not clearly recorded. The data is heavily skewed and does not give us clear reasons for the cause of the most common crashes.

###  `Yearly Accident Trends`

This graph shows how number of accidents per year changes over time.

# Significance of Reverse Geocoding Top Crash Locations

This section highlights the importance of converting geographic coordinates into meaningful location information to better understand traffic incidents:

1. **Identifying Hotspots**  
By combining latitude and longitude into a unique key and counting crashes per location, we can pinpoint areas with the highest frequency of incidents. This helps authorities focus interventions where they are needed most.

2. **Human-Readable Locations**  
Coordinates alone are not intuitive for analysis or reporting. Reverse geocoding transforms these numeric values into real-world addresses, making it easier for planners, policymakers, and stakeholders to interpret the data.

3. **Data-Driven Decision Making**  
Knowing the exact locations of high-crash areas allows city planners, traffic enforcement agencies, and safety teams to implement targeted safety measures, such as better signage, traffic signals, or road maintenance.

4. **Handling Data Limitations**  
The code includes mechanisms to handle invalid or missing coordinates and avoids overwhelming external geocoding services through rate limiting. This ensures reliable and responsible use of external APIs.

5. **Insight for Strategic Planning**  
By linking crash frequency with precise locations, this process supports evidence-based planning, helps in prioritizing resources, and enables detailed reporting for urban safety improvement initiatives.



```python
# First, install the required package
!pip install geopy

# Import necessary libraries
import pandas as pd
import time
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Your data dictionary
data = {
    'Location_Key': [
        '41.7514606032,-87.585971993',
        '41.7914202821,-87.5801477687',
        '41.722257273,-87.5852755651',
        '41.7546601244,-87.7413847586',
        '41.8808560477,-87.6176358918',
        '41.7893293233,-87.7416456442',
        '41.8968049724,-87.6170274232',
        '41.900752974,-87.6242349998'
    ],
    'Number_of_Crashes': [1610, 884, 686, 685, 534, 439, 393, 384]  # Note: I adjusted the length to match Location_Key
}
top_locations_df = pd.DataFrame(data)

# Defining the Reverse Geocoding Function 

# Initializing geocoder (using OpenStreetMap Nominatim, it requires a user_agent)
geolocator = Nominatim(user_agent="accident_analysis_project")

# Defining a function to look up the location name from the coordinate key
def get_location_name(coord_key):
    try:
        # 1. Split the key into latitude and longitude
        lat, lon = map(float, coord_key.split(','))

        # 2. Perform the reverse geocoding lookup
        # We use a short timeout to prevent the call from hanging indefinitely
        location = geolocator.reverse((lat, lon), timeout=5)

        # 3. Return the address, using .address for the full string
        if location:
            return location.address
        else:
            return 'Address Not Found'
    except Exception as e:
        # Catch errors like connection timeouts or invalid coordinates
        return f'Geocoding Failed: {e}'

# Wrap the function in RateLimiter to avoid overwhelming the external service
# Nominatim policy requires a delay between requests
geocode_limited = RateLimiter(get_location_name, min_delay_seconds=1.5)

#  Apply the Function to the DataFrame ---
start_time = time.time()

# Apply the rate-limited function to the Location_Key column
top_locations_df['Generated_Location_Name'] = top_locations_df['Location_Key'].apply(geocode_limited)

end_time = time.time()
#PRINTING RESUTS
# Display the key, the new generated name, and the crash count
print(top_locations_df[['Generated_Location_Name', 'Number_of_Crashes', 'Location_Key']])
```

To address class imbalance in our target variable `Cause_Group`, we applied Random OverSampling using RandomOverSampler from the imblearn library. This technique randomly duplicates samples from minority classes to ensure a more balanced class distribution, which can help improve model performance and reduce bias toward majority classes.

This section performs **class balancing** on the dataset to address class imbalance, which is common in classification problems where some categories appear much more frequently than others.  

## Separating Features and Target
- `X` contains all predictor variables (features).  
- `y` contains the target variable `Cause_Group`, which is the class to predict.  

## Initializing the Oversampler
- `RandomOverSampler` from **imbalanced-learn** is used to oversample minority classes.  
- `sampling_strategy='auto'` ensures all minority classes are oversampled to match the majority class.  
- `random_state=42` ensures reproducibility.  

## Resampling the Data
- `fit_resample(X, y)` generates a new balanced dataset.  
- `X_resampled` contains the features with oversampled rows.  
- `y_resampled` contains the corresponding balanced target labels.  

## Significance
- Balancing the classes prevents models from being biased toward majority classes.  
- Improves performance metrics like **recall** and **F1-score** for minority classes.  
- Random oversampling is simple and effective, but may increase **overfitting**.  
- More advanced techniques like **SMOTE** can create synthetic samples for better generalization.  

This step ensures that the predictive model can learn equally from all classes and not ignore minority categories.


```python
# performing class balancing using SMOTE in attempt to improve model performance for our df

from imblearn.over_sampling import RandomOverSampler

X = df2.drop('Cause_Group', axis=1)
y = df2['Cause_Group']
ros = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)
```

```python
# separating our target variable and features
X_resampled = df2.drop('Cause_Group', axis=1)
y_resampled = df2['Cause_Group']
```

```python
# Split into training + temp (temp will later be split into val + test)
X_resampled_train, X_resampled_temp, y_resampled_train, y_resampled_temp = train_test_split(
    X_resampled, y_resampled, test_size=0.3, random_state=42, stratify=y_resampled
)
```

# The following plot displays the ROC curves for each class in the Cause_Group target variable. These curves illustrate the trade-off between the True Positive Rate (TPR) and False Positive Rate (FPR) at various classification thresholds.

# Random Forest Model: ROC Curve for Multi-Class Classification

This section visualizes the Receiver Operating Characteristic (ROC) curves for each class in the Random Forest classifier:

- **Binarization:** The multi-class target is binarized to allow ROC computation for each class separately.
- **Predicted Probabilities:** The classifier’s predicted probabilities for each class are used to calculate the True Positive Rate (TPR) and False Positive Rate (FPR).
- **ROC & AUC:** For every class, an ROC curve is plotted, and the Area Under the Curve (AUC) is computed. Higher AUC values indicate better separability and prediction performance.
- **Visualization:** The ROC curves allow comparison of class-wise performance, highlighting which classes the model predicts confidently and which are more challenging. The diagonal line represents random guessing; curves above it indicate better-than-random predictions.


```python
import matplotlib.pyplot as plt
from sklearn.metrics import auc, roc_curve
from sklearn.preprocessing import label_binarize

# Binarize the output for multi-class ROC
y_test_bin_rf = label_binarize(y_resampled_test, classes=rf_clf.classes_)
n_classes_rf = y_test_bin_rf.shape[1]

# Get predicted probabilities (shape: [n_samples, n_classes])
y_prob_rf = rf_clf.predict_proba(X_resampled_test)

# Compute ROC curve and ROC area for each class
fpr_rf = dict()
tpr_rf = dict()
roc_auc_rf = dict()

for i in range(n_classes_rf):
    fpr_rf[i], tpr_rf[i], _ = roc_curve(y_test_bin_rf[:, i], y_prob_rf[:, i])
    roc_auc_rf[i] = auc(fpr_rf[i], tpr_rf[i])

# Plot the visualization
plt.figure()
for i in range(n_classes_rf):
    plt.plot(fpr_rf[i], tpr_rf[i], label='Class %d (AUC = %0.2f)' % (i, roc_auc_rf[i]))
    
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - Random Forest')
plt.legend(loc="lower right")
plt.show()
```

# NEURAL NETWORK

Let's build a neural network as our final model.

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, accuracy_score, f1_score
```

Let's use the copy we had already cleaned and label encoded to split our data into the train and test split


## Data Preparation for Neural Network

- **Separate Features and Target**: 
  - `X` contains all independent variables.  
  - `y` contains the dependent variable `Cause_Group`.

- **Train-Test Split**: 
  - The dataset is divided into training and testing sets (80% train, 20% test).  
  - Stratification preserves the class distribution in both sets.

- **Feature Scaling**: 
  - `StandardScaler` standardizes all features to a similar range.  
  - This improves neural network convergence and performance.

- **Target Encoding**: 
  - Multi-class target variable is converted to one-hot encoded format using `to_categorical`.  
  - This allows the neural network to output class probabilities correctly.


```python
#divide into dependent and independent variables
X = df2.drop('Cause_Group', axis=1)
y = df2['Cause_Group']

# split into train and test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=10, stratify=y
)

#let us scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Encode the target variable for multi-class classification
num_classes = len(np.unique(y_train))
y_train_encoded = to_categorical(y_train, num_classes=num_classes)
y_test_encoded = to_categorical(y_test, num_classes=num_classes)


```

## Neural Network Model Training

- **Model Architecture**:
  - **Input Layer**: Fully connected (`Dense`) layer with 128 neurons, using ReLU activation.
  - **Dropout Layer**: 30% dropout to reduce overfitting.
  - **Hidden Layer**: Fully connected layer with 64 neurons, ReLU activation.
  - **Dropout Layer**: Another 30% dropout.
  - **Output Layer**: `Dense` layer with neurons equal to the number of classes, using softmax activation for multi-class classification.

- **Compilation**:
  - Optimizer: `Adam` for efficient gradient-based optimization.
  - Loss Function: `categorical_crossentropy` for multi-class targets.
  - Metric: `accuracy` to track performance during training.

- **Training**:
  - Trained on scaled training data with one-hot encoded labels.
  - 10 epochs and batch size of 64.
  - 20% of training data used for validation to monitor overfitting.


```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)


history = model.fit(
    X_train_scaled,
    y_train_encoded,
    epochs=10,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)
```

## Neural Network Model Evaluation

After training the neural network, its performance is assessed on the test set to determine how well it generalizes to unseen data. Key metrics include:

## Enhanced Neural Network Training

In this improved neural network:

- **Architecture Upgrades**: Added more neurons in the hidden layers and included Batch Normalization to stabilize and speed up training, while Dropout layers reduce overfitting.  
- **Optimizer Tuning**: The Adam optimizer is used with a lower learning rate (0.0005) to allow more precise weight updates, improving convergence.  
- **Early Stopping**: Monitors validation loss and stops training if no improvement is observed for several epochs, preventing overfitting and saving the best model.  
- **Training


```python
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping


#Improved Neural Network

model = Sequential([
    Dense(256, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    BatchNormalization(),
    Dropout(0.4),

    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),

    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),

    Dense(num_classes, activation='softmax')
])

#  Compile model with lower learning rate 
optimizer = Adam(learning_rate=0.0005)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# --- Early stopping callback ---
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)


history = model.fit(
    X_train_scaled,
    y_train_encoded,
    epochs=50,
    batch_size=64,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

```

## Model Evaluation – Improved Neural Network

After training the enhanced neural network, we evaluated how well it performs on unseen data:

- **Test Accuracy**  
  Measuring accuracy on the test set helps determine the model’s generalization ability. A higher accuracy means the network can correctly classify a larger proportion of accident cause groups it has never seen before.

- **Predictions vs. True Labels**  
  The predicted class labels are compared against the actual labels to understand how often the model makes correct decisions for each category.

- **Classification Report**  
  This provides precision, recall, and F1-scores for every class.  
  Precision shows how often predicted classes are right, recall shows how well the model identifies all true cases, and the F1-score balances both. This breakdown is crucial when dealing with imbalanced target classes.

- **Macro F1-score**  
  The macro average treats every class equally, regardless of frequency.  
  This value is especially useful here because some crash cause categories appear far less often. A higher macro F1 means the model is more balanced and not just performing well on dominant classes.

Taken together, these metrics give a holistic picture of how the neural network handles multi-class classification beyond raw accuracy, especially important for real-world accident prediction where minority causes still matter.


```python
loss, accuracy = model.evaluate(X_test_scaled, y_test_encoded)
print(f"\nTest Accuracy: {accuracy:.4f}")

# Predict classes
y_pred = np.argmax(model.predict(X_test_scaled), axis=1)
y_true = np.argmax(y_test_encoded, axis=1)

# F1-score and classification report
print("\nClassification Report:\n", classification_report(y_true, y_pred))
print("\nMacro F1-score:", f1_score(y_true, y_pred, average='macro'))
```

## Interpreting Neural Network Predictions for Accident Causes

This snippet demonstrates how to convert the output of a neural network into a human-readable accident cause.

1. **Prediction:** The trained model (`loaded_model`) predicts probabilities for each class given the input features.

2. **Class Selection:** `np.argmax` identifies the index of the class with the highest probability, representing the predicted accident cause.

3. **Mapping to Labels:** A series of conditional checks (`if-elif`) converts the numeric index into a descriptive label such as:
   - "Speed & Right-of-Way Failure"
   - "Maneuver & Positioning Error"
   - "Distraction & Visibility"
   - … and so on.

4. **Output:** The final printed result tells the user the predicted primary cause of the accident, making the model's output easy to understand and actionable.


```python
prediction = loaded_model.predict(input_data_reshaped)
predicted_class = np.argmax(prediction[0])  # Extract the index of highest probability

if predicted_class == 0:
    print('Speed & Right-of-Way Failure')
elif predicted_class == 1:
    print('Maneuver & Positioning Error')
elif predicted_class == 2:
    print('Disregard Traffic Controls')
elif predicted_class == 3:
    print('Impairment & Recklessness')
elif predicted_class == 4:
    print('Distraction & Visibility')
elif predicted_class == 5:
    print('External/Environmental/Mechanical')
elif predicted_class == 6:
    print('Unclassified/Administrative')
elif predicted_class == 7:
    print('UNABLE TO DETERMINE')
```
